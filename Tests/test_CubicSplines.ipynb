```python
import PyKorg
import numpy as np
```

```python
PyKorg.CubicSplines.CubicSpline(1,2,3,4,True).show()
```

```python
xs = np.linspace(0,5,51)
ys = np.sin(xs)
```

```python
testCS = PyKorg.CubicSplines.construct_spline(xs,ys)
testCS.show()
```

```python
t1 = testCS.interpolate(1.35) 
v1 = np.sin(1.35)
print(t1,"vs",v1)
print("error:",(t1-v1)/v1)
print("normalized:",t1/v1)
```

```python
testCS.interpolate(.1) # true value is 0.0998334166468 (this is testing inputting a given value)
```

```python
t2 = testCS.cumulative_integral(1,2) 
v2 = 0.956449142415 # true value
print(t2[-1],"vs",v2)
print("error:",(t2[-1]-v2)/v2)
print("normalized:",[x/v2 for x in t2])
```
